# Audios de la sonda espacial Cassini
La sonda Cassini a su paso por Saturno, registro audios de algunos fenomenos que se presentan en el planeta. Estos sonidos tienen un patron claro y unas frecuencias caracterisiticas con las cuales se busca hacer una musicalización, para ello partimos de identificar estas frecuencias y dicho patron mediante un análisis frecuencial por medio de la transformada rapida de fourier.

Se análizaran seis audios (de donde se tomaron y que son cada uno)

```python
import matplotlib.pyplot as plt  #visualización grafica 
from scipy import signal # Espectograma 
from scipy.io import wavfile # Cargar .wav
import pandas as pd # Manejo de datos
import numpy as np  # Manejo de datos
import cass
```

```python
### Bizarre Features of Saturn's Radio Emissions (descripción)
```

```python
#cargar el audio
sample_rate, samples =  wavfile.read('Audios/Bizarre Features of Saturn\'s Radio Emissions.wav')
t,f,Sxx = cass.spectrograma2D(samples,sample_rate,1,"2D","gnuplot2")
```
